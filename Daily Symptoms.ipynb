```python
df = pd.read_csv('column_names.csv', header = None)
column_names = df.iloc[0].tolist()
string_values = column_names[:7]
date_values = [column_names[7]]
float_values =  column_names[8:]
```

```python
string_fields = []
for word in string_values:
    string_fields.append(StructField(word, StringType(), True))

date_fields = []
for word in date_values:
    date_fields.append(StructField(word, DateType(), True))

float_fields = []
for word in float_values:
    float_fields.append(StructField(word, FloatType(), True))
    
struct_fields = string_fields + date_fields + float_fields
customSchema = StructType(struct_fields)
```

```python
df = spark.read.load('hdfs://localhost:9000/user1/2021_country_daily_2021_IE_daily_symptoms_dataset.csv', format="csv", header="true", sep=',', schema=customSchema)
```
